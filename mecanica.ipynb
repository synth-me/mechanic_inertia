```python
from math import *
import pandas as pd
import re

def frequencia(periodo):
        
    f = 1/periodo   
        
    return f 

def frequencia_angular(frequencia):
    
    w = frequencia*2*pi
    
    return w 

def raio_de_giração_kg(d,w):

    f1 = (d*9.8)/(w**2)
    f2 = d**2


    kg = sqrt((f1-f2))
    return kg 

def raio_de_giração_k0(d,w):

    k0 = sqrt((9.8*d))/w

    return k0

def momento_inércia_i0(m,k):

    momento = m*(k**2)

    return momento  

def erro_percentual(inicial,final):
    
    a = final - inicial 
    
    if a > 0 : 
        pass
        s = a
    else:
        s = a*-1
    
    p = (s/final)*100
    
    return str(round(p,3))+'%' 

def momento_disco(r,m):
    
    x = r**4
    
    n = (pi*x)/4
    
    return n 

def momento_retangulo(d,m):
    
    n = ((d)*(d**3))/12
    
    return n 

def momento_barra(d,m):
    
    n = (1/3)*((m*((2*d)**2)))
    
    return n 

def momento_boneco(d,m):
    return d

info = {

"disco(teórico)":{
    "d(m)":0.00475,
    "massa(kg)":0.02,
    "frequencia(Hz)":2.28,
    "frequencia ang(rad/s)":14.36,
    "raio de gir k0(m)":0,
    "raio de gir kG(m)":2,
    },
"retangulo(teórico)":{
    "d(m)":0.018,
    "massa(kg)":0.015,
    "frequencia(Hz)":1.17,
    "frequencia ang(rad/s)":7.35,
    "raio de gir k0(m)":1,
    "raio de gir kG(m)":1
    },
"barra(teórico)":{
    "d(m)":0.028,
    "massa(kg)":0.115,
    "frequencia(Hz)":0.94,
    "frequencia ang(rad/s)":5.91,
    "raio de gir k0(m)":1,
    "raio de gir kG(m)":1
    },
"boneco(teórico)":{
    "d(m)":0.016,
    "massa(kg)":0.045,
    "frequencia(Hz)":1.24,
    "frequencia ang(rad/s)":7.82,
    "raio de gir k0(m)":1,
    "raio de gir kG(m)":1
    }
}

info_2 = {
    "disco":{
        "d(m)":0.00475,
        "massa(kg)":0.02,
        "periodo(s)":0.5
    },
    "retangulo":{
        "d(m)":0.018,
        "massa(kg)":0.015,
        "periodo(s)":0.75
    },
    "barra":{
        "d(m)":0.028,
        "massa(kg)":0.115,
        "periodo(s)":1
    },
    "boneco":{
        "d(m)":0.016,
        "massa(kg)":0.045,
        "periodo(s)":0.75
    }
}

formulas = {
    'disco(teórico)':momento_disco,
    'retangulo(teórico)':momento_retangulo,
    'barra(teórico)':momento_barra
}

for i in info:
    info[i]["raio de gir k0(m)"] = raio_de_giração_k0(info[i]['d(m)'],info[i]["frequencia ang(rad/s)"])
    info[i]['raio de gir kG(m)'] = raio_de_giração_kg(info[i]['d(m)'],info[i]["frequencia ang(rad/s)"])
    info[i]['momento de inercia i0(kg*m^2)'] = momento_inércia_i0(info[i]['massa(kg)'],info[i]["raio de gir k0(m)"])

for j in info_2: 
    info_2[j]['frequencia(Hz)'] = frequencia(info_2[j]["periodo(s)"])
    info_2[j]["frequencia ang(rad/s)"] = frequencia_angular(info_2[j]['frequencia(Hz)'])
    info_2[j]["raio de gir k0(m)"] = raio_de_giração_k0(info_2[j]['d(m)'],info_2[j]["frequencia ang(rad/s)"])
    info_2[j]['raio de gir kG(m)'] = raio_de_giração_kg(info_2[j]['d(m)'],info_2[j]["frequencia ang(rad/s)"])
    info_2[j]['momento de inercia i0(kg*m^2)'] = momento_inércia_i0(info_2[j]['massa(kg)'],info_2[j]["raio de gir k0(m)"])
    del info_2[j]['periodo(s)']
    
a = pd.DataFrame(info_2)      

z = {**info,**info_2}

error = {
    "erro % disco":{
        "d(m)":0,
        "massa(kg)":0,
        "frequencia(Hz)":0,
        "frequencia ang(rad/s)":0,
        "raio de gir k0(m)":0,
        'raio de gir kG(m)':0,
        'momento de inercia i0(kg*m^2)':0
    },
    "erro % retangulo":{
        "d(m)":0,
        "massa(kg)":0,
        "frequencia(Hz)":0,
        "frequencia ang(rad/s)":0,
        "raio de gir k0(m)":0,
        'raio de gir kG(m)':0,
        'momento de inercia i0(kg*m^2)':0
    },
    "erro % barra":{
        "d(m)":0,
        "massa(kg)":0,
        "frequencia(Hz)":0,
        "frequencia ang(rad/s)":0,
        "raio de gir k0(m)":0,
        'raio de gir kG(m)':0,
        'momento de inercia i0(kg*m^2)':0
    },
    "erro % boneco":{
        "d(m)":0,
        "massa(kg)":0,
        "frequencia(Hz)":0,
        "frequencia ang(rad/s)":0,
        "raio de gir k0(m)":0,
        'raio de gir kG(m)':0,
        'momento de inercia i0(kg*m^2)':0
    }
}

pairs = {
    "erro % disco":0,
    "erro % retangulo":0,
    "erro % barra":0,
    "erro % boneco":0
}

for k in list(z.keys()):
    base = str(k)
    for n in list(z.keys()):
        if re.findall(str(n),base) and n == base :
            pass
        elif re.findall(str(n),base):
            p = (n,base)
            for pr in pairs.keys():
                if re.findall(n,pr):
                    pairs[pr] = p 
        else:
            pass 
                
for e in error :
    pe0 = pairs[e][0]
    pe1 = pairs[e][1]
    
    error[e]['d(m)'] = erro_percentual(z[pe0]['d(m)'],z[pe1]['d(m)'])
    error[e]['massa(kg)'] = erro_percentual(z[pe0]['massa(kg)'],z[pe1]['massa(kg)'])
    error[e]["frequencia(Hz)"] = erro_percentual(z[pe0]["frequencia(Hz)"],z[pe1]["frequencia(Hz)"])
    error[e]["frequencia ang(rad/s)"] = erro_percentual(z[pe0]["frequencia ang(rad/s)"],z[pe1]["frequencia ang(rad/s)"])
    error[e]["raio de gir k0(m)"] = erro_percentual(z[pe0]["raio de gir k0(m)"],z[pe1]["raio de gir k0(m)"])
    error[e]['raio de gir kG(m)'] = erro_percentual(z[pe0]['raio de gir kG(m)'],z[pe1]['raio de gir kG(m)'])
    error[e]['momento de inercia i0(kg*m^2)'] = erro_percentual(z[pe0]['momento de inercia i0(kg*m^2)'],z[pe1]['momento de inercia i0(kg*m^2)'])

w = {**info,**info_2,**error}
wp = pd.DataFrame(w)
wp
```
